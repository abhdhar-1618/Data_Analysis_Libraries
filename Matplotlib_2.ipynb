### need to find a way to randomly generate numbers in x and y variables where numbers generated in x and y are not equal and when plotted, it draws a straight line

```python
#Labling

x = list(range(1, 11))
y = list(range(1, 11))
random.shuffle(x)
random.shuffle(y)
x.sort() 
y.sort()
plt.plot(x, y)
plt.title("First graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Labling and font change

x = list(range(1, 11))
y = list(range(1, 11))
random.shuffle(x)
random.shuffle(y)
x.sort() 
y.sort()
plt.plot(x, y)
plt.title("First graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

x = list(range(1, 21))
y = list(range(1, 21))
random.shuffle(x)
random.shuffle(y)
x.sort() 
y.sort()

plt.plot(x, y, 'b^--', label="First graph")

# Line # 2
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')  # Corrected line

plt.title("First graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

x = list(range(1, 21))
y = list(range(1, 21))
random.shuffle(x)
random.shuffle(y)
x.sort() 
y.sort()

#Graph resizing

plt.figure(figsize=(5,3), dpi=200)

plt.plot(x, y, 'b^--', label="First graph")

# Line # 2
x2 = np.arange(0, 4.5, 0.5)
x2.sort()
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')  # Corrected line

plt.title("First graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend()
plt.show()
```
