```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Generating random data from numpy library

X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
plt.scatter(X_data, y_data);
```

```python
plt.scatter(X_data, y_data, c="red");
```

```python
plt.scatter(X_data, y_data, c="red", marker="*");
```

```python
# "alpha" command is for transparency

plt.scatter(X_data, y_data, c="red", marker="*", alpha=0.4);
```

```python
years = [2006 + x for x in range(16)]
weights = [random.randint(80, 100) for _ in range(16)]

print(weights)
print(years)
```

```python
plt.plot(years, weights);
```

```python
plt.scatter(years, weights);
```

```python
plt.plot(years, weights, c="red");
```

```python
# for a thicker line

plt.plot(years, weights, c="green", lw=8);
```

```python
# for a thicker and dashed line

plt.plot(years, weights, c="green", lw=8, linestyle="--");
```

```python
plt.plot(years, weights, "r--", lw=3);
```
